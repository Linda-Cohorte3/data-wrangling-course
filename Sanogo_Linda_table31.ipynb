```python
df = pd.read_excel(source, header=5)
df.head(10)
```

```python
df = df.dropna(how="all", axis=1)
df.head(10)
```

```python
df = df.dropna(how="all", axis=0).reset_index(drop=True)
df.head(10)
```

```python
columns=[]
for t in df.columns:
     columns.append([str(j) for j in df[t][:6] if pd.notnull(j)])           
columns
```

```python
col=["HDI_rank",
     "Countries",
     "FEA_Rate_%_2005",
     "FEA_Index (1990=100) 2005",
     "FEA_%_male_rate 2005",
     "EEA_%_Agri_Women 1995-2005b",
     "ref1",
     "EEA_%_Agri_Men 1995-2005b",
     "ref2",
     "EEA_%_Ind_Women 1995-2005b",
     "ref3",
     "EEA_%_Ind_Men 1995-2005b",
     "ref4",
     "EEA_%_Ser_Women 1995-2005b",
     "ref5",
     "EEA_%_Ser_Men 1995-2005b",
     "ref6",
     "CFW_%_Women 1995-2005b",
     "CFW_%_Men 1995-2005b"
         ]
```

```python
df.columns=col
df.head(10)
```

```python
df=df.drop(np.arange(6),axis=0).reset_index(drop=True)
df.head(20)
```

```python
df.tail(30)
```

```python
df=df.drop(np.arange(180,len(df)),axis=0).reset_index(drop=True)
df.tail(30)
```

```python
hdi_categories = ["HIGH HUMAN DEVELOPMENT", "MEDIUM HUMAN DEVELOPMENT", "LOW HUMAN DEVELOPMENT"]
df=df.drop(df[df["HDI_rank"].isin(hdi_categories)].index,axis=0).reset_index(drop=True)
df.head()
```

```python
df.loc[:, "HDI_category"] = np.where(df["HDI_rank"] > 155,
                                     "LOW HUMAN DEVELOPMENT",
                                     np.where(df["HDI_rank"] < 71,
                                              "HIGH HUMAN DEVELOPMENT",
                                              "MEDIUM HUMAN DEVELOPMENT"
                                              
                                             )
                                    )
df.sort_values(by="HDI_rank", inplace=True)
df.head()
```

```python
df.to_csv(index=False)
```
