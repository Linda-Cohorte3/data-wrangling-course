```python
#importation de whyqd et appelle de la methode schema
import whyqd as _w
schema = _w.Schema()
```

```python
#ajout des details a ton schema
details = {
    "name":"HDR_Table_31_Schema",
    "title":"Gender inequality in economic activity_HDR 2007-2008 Table 31",
    "description":" schema for exercice number one"
    
}
schema = _w.Schema()

schema.set_details(**details)
```

```python
fields=[
        {
            "name":"Country Name",
            "title":"Country Name",
            "type":"string",
            "description":"Official country names",
            "constraints": {
                "required":True
            }
        },    
        {
            "name":"HDI Category",
            "title":"HDI Category",
            "type":"string",
            "description":"Human Development Index Category derived from the HDI Rank",
            "constraints":{
                "required": True
             }
           
        },
        {
            "name":"Indicator Name",
            "title":"Indicator Name",
            "type":"string",
            "description":"Indicator described in the data series"
        },
        {
            "name":"Values",
            "title":"Values",
            "type":"number",
            "description":"Value for the Year and Indicator Name"

        }
    ]
```

```python
#ajout de fiels dans ton schema
for field in fields:
    schema.set_field(**field)
```

```python
directory = r"C:/Users/La licorne/Desktop/data-wrangling-course/" #je fais appel à mon repertoire courant
filename = "Lin31" # je nomme mon fichier
# le fichier existe déjà alors je spécifie "overwrite=True" pour ne pas avoir d'erreur
schema.save(directory, filename=filename, overwrite=True)
```

```python
### Les importations et paramètres suivants vous permettent d'obtenir un large éventail de résultats pour vos tableaux
from IPython.core.display import HTML
display(HTML("<style>pre { white-space: pre !important; }</style>"))

import numpy as np
import whyqd as _w

SCHEMA_SOURCE = "lin31.json" #"lin31.json" est le nom de mon schema créé ci haut
DIRECTORY = directory
INPUT_DATA = [
    "HDR 2007-2008 Table 31.xlsx"  #"HDR 2007-2008 Table 31.xlsx" c'est le nom de mon fichier inital de la leçon 1 1
]
method = _w.Method(SCHEMA_SOURCE, directory=DIRECTORY, input_data=INPUT_DATA)
```
